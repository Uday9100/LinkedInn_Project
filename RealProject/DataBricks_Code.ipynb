```
def get_job_alerts_by_skills(candidate_skills, job_skills_df, posting_data_df):
    # Join job_skills with posting_data to find matching jobs
    jobs_with_skills = job_skills_df.join(posting_data_df, "job_id", "inner")
    print("Jobs with Skills DataFrame:")
    jobs_with_skills.show(truncate=False)  # Show the entire DataFrame for better visibility
    
    # Filter for jobs matching the candidate's skills
    matching_jobs = jobs_with_skills.filter(jobs_with_skills.skill_abr.isin(candidate_skills))
    print("Matching Jobs DataFrame:")
    matching_jobs.show(truncate=False)  # Show the filtered jobs
    
    # Select relevant columns for alerts
    alerts = matching_jobs.select("company_name", "job_id", "location", "description")
    
    return alerts

# Test the function
candidate_skills = ["PRJM", "IT", "ENG"]  # Replace with the skills you want to match
alerts = get_job_alerts_by_skills(candidate_skills, job_skills_df, posting_data_df)

# Show alerts count and data
print("Alerts Count:", alerts.count())
alerts.show()
```

```
# Job Alerts Based on Location
def get_job_alerts_by_location(location, posting_data_df):
    # Print the input location for reference
    print("Looking for jobs in location:", location)
    
    # Filter postings by location
    location_jobs = posting_data_df.filter(posting_data_df.location == location)
    
    # Show the filtered DataFrame to verify the results
    print("Filtered Jobs by Location DataFrame:")
    location_jobs.show(truncate=False)  # Display all columns for better visibility
    
    # Select relevant columns for alerts
    alerts = location_jobs.select("company_name", "job_id", "location", "description")
    
    return alerts

# Test the function
location = "New York, NY"  # Replace with the desired location
alerts_location = get_job_alerts_by_location(location, posting_data_df)

# Show alerts count and data
print("Alerts Count:", alerts_location.count())
alerts_location.show()
```

```
# Trending Skills Analysis
def get_trending_skills(job_skills_df):
    # Count the number of postings for each skill
    trending_skills = job_skills_df.groupBy("skill_abr").count().orderBy("count", ascending=False)
    
    # Show the intermediate DataFrame to verify counts
    print("Trending Skills DataFrame (before limiting):")
    trending_skills.show(truncate=False)  # Display all columns for better visibility
    
    # Select the top N trending skills
    top_trending_skills = trending_skills.limit(10)
    
    # Show the top trending skills
    print("Top Trending Skills:")
    top_trending_skills.show(truncate=False)  # Display all columns for better visibility
    
    return top_trending_skills

# Test the function
trending_skills_df = get_trending_skills(job_skills_df)

# Show the count of trending skills
print("Count of Trending Skills:", trending_skills_df.count())
```

```
# Combining Alerts
def get_combined_alerts(candidate_skills, location, job_skills_df, posting_data_df):
    # Get job alerts by skills
    skill_alerts = get_job_alerts_by_skills(candidate_skills, job_skills_df, posting_data_df)
    
    # Show skill alerts
    print("Skill Alerts:")
    skill_alerts.show(truncate=False)  # Display all columns for better visibility

    # Get job alerts by location
    location_alerts = get_job_alerts_by_location(location, posting_data_df)
    
    # Show location alerts
    print("Location Alerts:")
    location_alerts.show(truncate=False)  # Display all columns for better visibility

    # Combine alerts into a single DataFrame
    combined_alerts = skill_alerts.union(location_alerts).distinct()
    
    # Show combined alerts
    print("Combined Alerts:")
    combined_alerts.show(truncate=False)  # Display all columns for better visibility

    return combined_alerts

# Test the function
candidate_skills = ["PRJM", "IT", "ENG"]  # Example skills to search for
location = "New York, NY"  # Example location to search for

combined_alerts_df = get_combined_alerts(candidate_skills, location, job_skills_df, posting_data_df)

# Show the count of combined alerts
print("Count of Combined Alerts:", combined_alerts_df.count())
```

```
# Skill Gap Analysis

def skill_gap_analysis(posting_data_df, job_skills_df, candidate_skills):
    # Get unique skills from job postings
    required_skills = job_skills_df.select("skill_abr").distinct()

    # Create a DataFrame of candidate skills
    candidate_skills_df = spark.createDataFrame([(skill,) for skill in candidate_skills], ["skill_abr"])

    # Find missing skills
    missing_skills = required_skills.join(candidate_skills_df, "skill_abr", "left_anti")

    return missing_skills

# Example usage
candidate_skills = ["PRJM", "IT", "ENG"]
missing_skills_df = skill_gap_analysis(posting_data_df, job_skills_df, candidate_skills)
print("Skills Needed to Meet Job Requirements:")
missing_skills_df.show()
```

```
#Job Recommendation System
def recommend_jobs_based_on_skills(candidate_skills, job_skills_df, posting_data_df):
    # Get job postings that require the candidate's skills
    jobs_with_candidate_skills = job_skills_df.filter(job_skills_df.skill_abr.isin(candidate_skills))\
        .join(posting_data_df, "job_id", "inner")\
        .select("job_id", "company_name", "location", "description")

    return jobs_with_candidate_skills

# Example usage
recommended_jobs_df = recommend_jobs_based_on_skills(candidate_skills, job_skills_df, posting_data_df)
print("Recommended Jobs Based on Skills:")
recommended_jobs_df.show(truncate=False)
```

```
#Job Alerts Based on User Preferences
def job_alerts_by_preferences(location, min_salary, max_salary, posting_data_df):
    # Filter jobs by user preferences
    job_alerts = posting_data_df.filter((posting_data_df.location == location) &
                                        (posting_data_df.min_salary >= min_salary) &
                                        (posting_data_df.max_salary <= max_salary))

    return job_alerts.select("job_id", "company_name", "location", "description")

# Example usage
location = "New York, NY"
min_salary = 60000
max_salary = 120000
job_alerts_df = job_alerts_by_preferences(location, min_salary, max_salary, posting_data_df)
print("Job Alerts Based on User Preferences:")
job_alerts_df.show(truncate=False)
```

```
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# Initialize Spark session (if not already done)
spark = SparkSession.builder.appName("Job Skills Analysis").getOrCreate()

# Function to print and verify if DataFrames are not empty
def check_df(df, name):
    print(f"{name} count: {df.count()}")
    df.show(5, truncate=False)  # Show the first 5 rows for inspection


# Check initial DataFrames
check_df(job_skills_df, "job_skills_df")
check_df(posting_data_df, "posting_data_df")
check_df(companies_df, "companies_df")

# Rename company_name column in posting_data_df to avoid ambiguity
posting_data_df = posting_data_df.withColumnRenamed("company_name", "posting_company_name")

# Select relevant fields from companies_df and rename to avoid ambiguity
# Remove 'value' since the columns are not nested
companies_df_cleaned = companies_df.select(
    F.col("company_id").alias("company_id"),
    F.col("name").alias("company_name"),
    F.col("description").alias("company_description")  # Renaming to avoid ambiguity
)

# Check the cleaned companies DataFrame
check_df(companies_df_cleaned, "companies_df_cleaned")

# Companies Hiring for Specific Skills
def companies_hiring_for_skills(candidate_skills, job_skills_df, posting_data_df, companies_df_cleaned):
    # Check if DataFrames have data
    check_df(job_skills_df, "job_skills_df")
    check_df(posting_data_df, "posting_data_df")
    check_df(companies_df_cleaned, "companies_df_cleaned")

    # Join job_skills_df with posting_data_df on job_id
    jobs_with_skills = job_skills_df.join(posting_data_df, "job_id", "inner")
    check_df(jobs_with_skills, "jobs_with_skills (after join with posting_data_df)")

    # Join the result with cleaned companies_df on company_id
    jobs_with_companies = jobs_with_skills.join(companies_df_cleaned, "company_id", "inner")
    check_df(jobs_with_companies, "jobs_with_companies (after join with companies_df_cleaned)")

    # Filter for jobs matching the candidate's skills
    matching_jobs = jobs_with_companies.filter(jobs_with_companies.skill_abr.isin(candidate_skills))
    check_df(matching_jobs, "matching_jobs (after filtering for skills)")

    # Select relevant columns and ensure distinct results
    companies_hiring = matching_jobs.select(
        F.col("company_name"),  # Explicitly reference the company_name from companies_df_cleaned
        "skill_abr",
        "location",
        F.col("company_description"),  # Using the renamed column
        F.col("posting_company_name")   # Select the renamed column from posting_data_df
    ).distinct()

    # Show companies hiring for specific skills
    print("Companies Hiring for the Specified Skills:")
    companies_hiring.show(truncate=False)

    return companies_hiring

# Example usage
candidate_skills = ["IT", "ENG", "PRJM"]  # Example skills

# Call the function and get the companies hiring for the specified skills
try:
    companies_hiring_df = companies_hiring_for_skills(candidate_skills, job_skills_df, posting_data_df, companies_df_cleaned)
except Exception as e:
    print(f"Error: {e}")
```
